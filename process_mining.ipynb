```python
def import_csv(file_path):
    event_log = pd.read_csv(file_path, sep=',')
    event_log = pm4py.format_dataframe(event_log, case_id='CaseID', activity_key='Activity', timestamp_key='Timestamp')
    start_activities = pm4py.get_start_activities(event_log)
    end_activities = pm4py.get_end_activities(event_log)
    print("Start activities: {}\nEnd activities: {}".format(start_activities, end_activities))

    return event_log.head(15)
```

```python
file_path = './complex_event_log_1000_cases.csv'
import_csv(file_path)
```

```python
# Visualize the processes
# https://pm4py.fit.fraunhofer.de/static/assets/api/2.7.8/api.html#visualization-pm4py-vis
net, initial_marking, final_marking = alpha_miner.apply(import_csv(file_path))
pn_visualizer(net, initial_marking, final_marking)
```
